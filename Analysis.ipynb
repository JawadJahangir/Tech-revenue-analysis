```python
import pandas as pd
import numpy as np

data = {
    "Year":[2018,2019,2020,2021,2022,2023],

    "Cloud_Services":[120,150,200,260,320,400],
    "AI_Solutions":[50,80,120,180,240,300],
    "Cybersecurity":[90,110,140,170,210,250],
    "Software_Products":[200,220,210,230,260,300]
}

df = pd.DataFrame(data)
df.to_csv("Dataset.csv")
```

```python
#Data Analysis (Pandas + NumPy)
#Total revenue per year
df["per_year_revenue"]=df.loc[:,["Cloud_Services","AI_Solutions","Cybersecurity","Software_Products"]].sum(axis=1)
print(df[["Year","per_year_revenue"]])
```

```python
#Average revenue of each product
average_revenue_Each_product=df.iloc[:,1:-2].mean()
print(average_revenue_Each_product)
```

```python
#Highest revenue product
total_product_revenue=df.loc[:,["Cloud_Services","AI_Solutions","Cloud_Services","Software_Products"]].sum()
higest_revune_product=total_product_revenue.idxmax()
print("Higest Revenue product:",higest_revune_product)
```

```python
import matplotlib.pyplot as plt
plt.plot(df["Year"],df["per_year_revenue"],marker="o",linestyle="dashed",color="yellow")
plt.title("Revenue Growth Per Year",fontsize=20,color="red")
plt.xlabel("Year",fontsize=15,fontweight="bold")
plt.ylabel("Revenue",fontsize=15,fontweight="bold")
plt.show()

```

```python
#Revenue Grwoth of All PRoduct Over Time
plt.plot(df["Year"],df["Cloud_Services"],marker="o",label="Cloud Services",linestyle="dashed")
plt.plot(df["Year"],df["AI_Solutions"],marker="o",label="AI Solutions",linestyle="-.")
plt.plot(df["Year"],df["Software_Products"],marker="o",label="Software Products",linestyle="--")
plt.plot(df["Year"],df["Cybersecurity"],marker="o",label="Cybersecurity")

plt.title("Revenue Growth of All Products",fontsize=20,color="red")
plt.xlabel("Year",fontsize=15,fontweight="bold")
plt.ylabel("Revenue",fontsize=15,fontweight="bold")
plt.legend(loc="upper left")
plt.show()
```

```python
#Show revenue comparison of products in 2023.

data_2023=df.iloc[5,1:-2]
plt.bar(data_2023.index,data_2023.values,color="green")
plt.title("Revenue Comparison of Products in 2023",fontsize=20)
plt.xlabel("Products",fontsize=15)
plt.ylabel("Revenue",fontsize=15)
for i in range(len(data_2023)):
       plt.text(i,data_2023.values[i]-20,str(data_2023.values[i]),ha="center",fontsize=10)
plt.show()
```

```python
#Horizontal Bar Show average revenue by product.
plt.barh(average_revenue_Each_product.index,average_revenue_Each_product.values,color="red")
plt.title("Average Revenue by Product",fontsize=20,color="blue")
plt.xlabel("Revenue",fontsize=15,fontweight="bold")
plt.ylabel("Products",fontsize=15,fontweight="bold")
plt.show()
```

```python
#Show market share of products in 2023.
plt.figure(figsize=(10,5))
explode=[0.2,0,0,0]
plt.pie(data_2023.values,
        labels=data_2023.index,
        autopct="%1.1f%%",
        textprops={"fontsize":10,"color":"gray"},
        explode=explode,
        startangle=380,
        shadow=True

        )
plt.legend(loc="upper left")
plt.title("Market Share of Products in 2023",fontsize=20,color="red",fontweight="bold")
plt.axis("equal")
plt.show()
```

```python
#Show revenue composition over years.
plt.figure(figsize=(10,5))
plt.stackplot(df["Year"],df["Cloud_Services"],df["AI_Solutions"],df["Cybersecurity"],df["Software_Products"],labels=["Cloud Services","AI Solutions","Cybersecurity","Software Products"])
plt.legend(loc="upper left")
plt.title("Revenue Composition Over Years",fontsize=20,color="red")
plt.xlabel("Year",fontsize=15,fontweight="bold")
plt.ylabel("Revenue",fontsize=15,fontweight="bold")
plt.grid(True)
plt.show()
```
